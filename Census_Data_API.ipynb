```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import requests
from us import states

# Census API Key
key_pd = pd.read_csv("Keys.py")
ckey = key_pd[key_pd['Name']== 'ckey']['key'].max().strip()
year=2015
```

```python
# Census Search to retrieve data on all zip codes (2016)
# Analyze population, race age, income relationship with store location

censusFields="NAME,B19013_001E,B01003_001E,B01002_001E,B19301_001E,B17001_002E,B02001_002E,B02001_003E,"
censusFields += "B02001_004E,B02001_005E,B02001_006E,B02001_007E,B02001_008E,B02001_009E,B02001_010E,B01001A_002E,"
censusFields += "B01001A_017E,B01001B_002E,B01001B_017E,B01001C_002E,B01001C_017E,B01001D_002E,B01001D_017E,B01001E_002E,"
censusFields += "B01001E_017E,B01001F_002E,B01001F_017E,B01001G_002E,B01001G_017E,B01001H_002E,B01001H_017E,B01001I_002E,B01001I_017E"


target_url="https://api.census.gov/data/%s/acs5?get=%s&for=zip+code+tabulation+area:*&key=%s" % (year,censusFields,ckey)
# print(target_url)

census_data = requests.get(target_url).json()

# # Convert to DataFrame
census_pd = pd.DataFrame(census_data)

census_pd.columns = census_pd.iloc[0]

# Column Reordering
census_pd = census_pd.rename(columns={"NAME": "Name",
                                      "B01003_001E": "Population", 
                                      "B01002_001E": "Median Age",
                                      "B19013_001E": "Household Income",
                                      "B19301_001E": "Per Capita Income",
                                      "B17001_002E": "Poverty Count",
                                      "B02001_002E": "Race White",
                                      "B02001_003E": "Race African American",
                                      "B02001_004E": "Race American Indian",
                                      "B02001_005E": "Race Asian",
                                      "B02001_006E": "Race Pacific Islander",
                                      "B02001_007E": "Race Some Other",
                                      "B02001_008E": "Race Two or More",
                                      "B02001_009E": "Race Two or More Including Some Other",
                                      "B02001_010E": "Race Two or More Excluding Some Other",
                                      "B01001A_002E": "Male White",
                                      "B01001A_017E": "Female White",
                                      "B01001B_002E": "Male African",
                                      "B01001B_017E": "Female African",
                                      "B01001C_002E": "Male Native American",
                                      "B01001C_017E": "Female Native American",
                                      "B01001D_002E": "Male Asian",
                                      "B01001D_017E": "Female Asian",
                                      "B01001E_002E": "Male Pacific Islander",
                                      "B01001E_017E": "Female Pacific Islander",
                                      "B01001F_002E": "Male Some Other Race",
                                      "B01001F_017E": "Female Some Other Race",
                                      "B01001G_002E": "Male Two or More Races",
                                      "B01001G_017E": "Female Two or More Races",
                                      "B01001H_002E": "Male White Not Hispanic",
                                      "B01001H_017E": "Female White Not Hispanic",
                                      "B01001I_002E": "Male Hispanic",
                                      "B01001I_017E": "Female Hispanic", "zip code tabulation area": "Postcode"})



# # drop first record
census_pd = census_pd.drop(census_pd.index[0])

# print(census_pd['Postcode'].dtype)
# print(zipcode_database_pd['Zipcode'].dtype)
```
